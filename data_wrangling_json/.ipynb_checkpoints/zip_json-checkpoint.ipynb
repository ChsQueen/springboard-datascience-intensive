```python
import json
```

```python
import pandas as pd
```

```python
from pandas.io.json import json_normalize
```

```python
import numpy as np
```

```python
with open("data/zips.json") as line:
    text = line.readlines()

jlist = []
for i in np.arange(len(text)):
    jlist.append({"id":json.loads(text[i])["_id"],
                 "city":json.loads(text[i])["city"],
                 "state": json.loads(text[i])["state"],
                 "pop": json.loads(text[i])["pop"],
                 "lat": json.loads(text[i])["loc"][0],
                 "lon": json.loads(text[i])["loc"][1]})
    
    #jdata = json.loads(str(text[0]))
jlist
```

```python
jlist
```

```python
dfj = json_normalize(jlist)
```

```python
dfj = dfj[["id", "city", "state", "pop", "lat", "lon"]]
```
