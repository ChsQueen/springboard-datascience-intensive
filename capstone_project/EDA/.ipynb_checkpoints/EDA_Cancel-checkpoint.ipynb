Let us now understand which factor affects the cancellation, if any. 

***Year***: Which year saw more cancellation rate?

```python
def plot_cancel_count_by_temporal(by, xlab, ylab, title, ylim):
    plt.subplots(1, 1, figsize=(10, 4))

    ax = plt.subplot(1, 1, 1)
    (100*dfs.groupby(by)["Cancelled"].sum() / dfs.groupby(by)["Cancelled"].count()).plot(kind="bar", 
                                                                                         color=tableau20[16])
    plt.xlabel(xlab)
    plt.ylabel(ylab)
    plt.ylim(ylim)
    plt.title(title)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
```

```python
plot_cancel_count_by_temporal("Year", "Year", "% Cancellation", "% Cancellation by year", [0, 1.75])
```

We see an improvement in 2016 from previous year. Is there any partucular quarter in which most flights are cancelled?

***Quarter***

```python
plot_cancel_count_by_temporal("Quarter", "Quarter", "% Cancellation", "% Cancellation by quarter", [0, 2.6])
```

The highest rate of cancellation (about 2.5%) is seen in the first quarter. The other quarters have lower and quite similar cancellation rate, about 1%. The highest rate in the first quarter may be due to the winter storm. To refine this observation, let us see the monthly cancellation rates.


***Month***

```python
plot_cancel_count_by_temporal("Month", "Month", "% Cancellation", "% Cancellation by month", [0, 3.5])
```

Clearly, January and Febreuary have the highest cancellation rate, corroborating with previous quarterly chart. We also see that the lowest rate of cancellation is observed in September. 


How about ***Day of month***?

```python
plot_cancel_count_by_temporal("DayofMonth", "Day of month", "% Cancellation", "% Cancellation by day of month", [0, 2.5])
```

There seems to be more cancellation in the third week of the month, however the trend is not very clear. We may omit this factor.


***Day of week***

```python
plot_cancel_count_by_temporal("DayOfWeek", "Day of week", "% Cancellation", "% Cancellation by day of week", [0, 2])
```

The lowest cancellation rate (about 1.2%) is seen beetween Wednesday to Friday. From Saturday onwards the rate increases and peaks on Monday to about 1.7%, and then drops on coming days.

Since the number of unique cities are 322, we will end up with quite complicated model if we add this variable in the model. To simply, we might coarse-grain the airport location information by considering the OriginState variable.

```python
def plot_cancel_count_by_spatial(by, xlab, ylab, title, ylim, sort):
    fig, ax = plt.subplots(1, 1, figsize=(20, 12))
    can_rate = (100*dfs.groupby(by)["Cancelled"].sum() / dfs.groupby(by)["Cancelled"].count())
    can_rate_sorted = can_rate.sort_values(ascending=False)
    if sort:
        ax = can_rate_sorted.plot(kind='bar',facecolor=tableau20[16])
    else:
        ax = can_rate.plot(kind='bar',facecolor=tableau20[16])
    plt.xlabel(xlab)
    plt.ylabel(ylab)
    plt.ylim(ylim)
    plt.title(title)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    plt.setp( ax.xaxis.get_majorticklabels(), rotation=90)
    #ax.invert_yaxis()
```

```python
dfsr = dfs[["Quarter", "Month", "DayOfWeek", "UniqueCarrier", "OriginState", "DestState", 
           "CRSDepHr", "CRSArrHr", "DistanceGroupNew", "Cancelled"]]
dfsr = dfsr.ix[np.random.choice(dfs.index.values, 100000, replace=False)]
dfsr.shape
```

For some categorical variabes, we need to create additional indicator features. 

```python
dfsr = pd.concat([dfsr, pd.get_dummies(dfsr.UniqueCarrier, prefix="UniqueCarrier")], axis=1)
dfsr.drop(["UniqueCarrier"], inplace=True, axis=1)
```

```python
dfsr.shape
```

```python
dfsr = pd.concat([dfsr, pd.get_dummies(dfsr.OriginState, prefix="OriginState")], axis=1)
dfsr.drop(["OriginState"], inplace=True, axis=1)
```

```python
dfsr = pd.concat([dfsr, pd.get_dummies(dfsr.DestState, prefix="DestState")], axis=1)
dfsr.drop(["DestState"], inplace=True, axis=1)
```

```python
dfsr.shape
```

We now use PCA to better understand which predictor variable explains more variance. However, before doing that we first split our data into train and test sets, and then normalize all our predictor variables.

***Splitting***

```python
X_train, X_test, y_train, y_test = train_test_split(dfsr.drop(["Cancelled"], axis=1), 
                                                    dfsr.Cancelled, test_size=0.5, random_state=10)
```

***Scaling***

```python
scaler = MinMaxScaler()
X_train_scale = scaler.fit_transform(X_train)
X_test_scale = scaler.transform(X_test)
```

***PCA***

```python
pca = PCA().fit(X_train_scale)
```

***Scree plot***

```python
plt.plot(pca.explained_variance_, c=tableau20[6])
plt.xlabel("Number of components")
plt.ylabel("Exaplained variance")
```

```python
plt.plot(pca.explained_variance_ratio_.cumsum(), c=tableau20[6])
plt.xlabel("Number of components")
plt.ylabel("Cumulative explained variance")
```

There is no distinct boundary between high and low values of explained variance, except last few components. We could go with 60-80 components to simplify the model little bit and loose the interpretability of variables. Let us not do that at this stage, and stick with all components, i.e. let us proceed without PCA components (or with original data set). Also, even if wanted to visualize first few components we cannot do that because to have even 50% of the variance we would need about 50 components! 2-3 first components represent only about 20% of the total varience. 

Let us now start modeling the data, with Logistic Regression (so we use the original data, i.e. non-scaled data)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import cohen_kappa_score
params = {"C": [0.01, 0.02, 0.05, 0.1]}
clf = GridSearchCV(LogisticRegression(class_weight = 'balanced'), cv=3,
                   param_grid=params, scoring='roc_auc')
clf.fit(X_train, y_train)
```

```python
clf.best_score_
```

```python
cohen_kappa_score(clf.predict(X_train), y_test)
```
